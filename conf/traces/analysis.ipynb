```python
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot(json_data, lower_bound=0, limit=None):
    x = []
    y = []
    if limit == None:
        limit = len(json_data['data'])
    for i in range(lower_bound, len(json_data['data'][:limit])):
        x.append(i)
        y.append(json_data['data'][i])
    plt.figure(figsize=(18,4))
    plt.plot(x,y)
    plt.show()
```
